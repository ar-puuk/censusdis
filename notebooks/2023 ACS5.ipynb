# 2023 ACS5

On December 12, 2024, the U.S. Census Bureau released the 
[ACS 5-Year Data](https://www.census.gov/data/developers/data-sets/acs-5year.html)
for 2023. This simple notebook demonstrates how `censusdis` was able to 
use the data immediately upon it's release.

```python
import censusdis.data as ced
from censusdis.datasets import ACS5
from censusdis.states import ALL_STATES_AND_DC
```

```python
ced.variables.search(ACS5, 2023, pattern="median household income")
```

```python
ced.download(ACS5, 2023, ["NAME", "B19013A_001E"], state="*")
```
